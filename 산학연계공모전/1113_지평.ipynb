```python
# 반려동물: 결측치 비율이 12%이지만, 대부분이 '없음'(없음 > 개 > 고양이 > 기타) -> '대체?
# 매니저사용휴대폰: 결측치 비율이 12%이며, 대부분이 '안드로이드' (안드로이드 > 기타 > ios) -> 최빈값인 '안드로이드'로 대체? or '기타'로 대체? 
# 매니저이동방법: 결측치 비율이 1%이며, 대부분이 '대중교통'이므로 최빈값인 '대중교통'으로 대체!
# CS교육이수여부: 결측치 비율이 12%이며, 대부분이 '미이수'이므로 최빈값인 '미이수'로 대체!
# 청소교육이수여부: 결측치 비율이 12%이며, 대부분이 '미이수'이므로 최빈값인 '미이수'로 대체!

imputer_col = ['반려동물', '매니저사용휴대폰', '매니저이동방법', 'CS교육이수여부', '청소교육이수여부']

for col in imputer_col:
    ftr.loc[ftr[col].isna(), col] = ftr[col].mode()[0]
    ftr_te.loc[ftr_te[col].isna(), col] = ftr_te[col].mode()[0]
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
def get_f15_score(y_test, pred):
    accuracy = accuracy_score(y_test, pred)
    precision = precision_score(y_test, pred)
    recall = recall_score(y_test, pred)
    f15 = ((1+1.5**2) * precision * recall) / ((1.5**2)*precision + recall)
    return f15
```

```python
# 하이퍼옵티마이저
seq_space = hp.choice('model',[
    {
        'units':hp.quniform('units', 16, 48, 8),
        'units1':hp.quniform('units1', 16, 128, 8),
        'units2':hp.quniform('units2', 16, 128, 8),
        'units3':hp.quniform('units3', 16, 128, 8),
        'activation':hp.choice('activation', ['mish', 'relu', 'elu', 'selu']),
        'batchsize':random.randint(32, 128),
        'dropout':random.uniform(0, 0.5),
        'dropout1':random.uniform(0, 0.5),
        'optimizer': hp.choice('optimizer',['Nadam', 'adam', 'rmsprop'])
    }
])
```

#### 모델 구축
- 많은 피쳐의 유의미한 특성만을 뽑기 위해 전체적인 모델의 구조를 Auto Encoding 방식에서 착안하였다.

```python
def objective(h):   
    input1 = Input(shape=(tr_cat.shape[1],))
    dense1 = layers.Dense(32,activation='relu')(input1)
    dense1 = layers.Dense(64,activation='relu')(dense1)

    input2 = Input(shape=(tr_num.shape[1],))
    dense2 = layers.Dense(32,activation='relu')(input2)
    dense2 = layers.Dense(64,activation='relu')(dense2)

    Add1 = layers.Concatenate()([dense1, dense2])
    output1 = layers.AlphaDropout(0.2)(Add1)
    output1 = layers.Dense(64, activation='selu')(output1)
    output1 = layers.Dense(h['units1'], activation='selu')(output1)
    output1 = layers.Dense(64, activation='selu')(output1)
    output1 = layers.Dense(128, activation='selu')(output1)

    model1 = Model([input1, input2], output1)

    model1.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['accuracy'])
    
    mc = tf.keras.callbacks.ModelCheckpoint('fin_Model.tf', monitor='val_accuracy', save_best_only=True)
    es = tf.keras.callbacks.EarlyStopping(monitor='val_accuracy', patience=3)
    
    model1.fit([tr_cat, tr_num], y_train, epochs=10, batch_size=h['batchsize'], validation_split=0.2, callbacks=[es, mc], verbose=0)
    
    fin_Model = load_model('fin_Model.tf')
    
    loss, acc = fin_Model.evaluate([val_cat, val_num], y_valid, verbose=0)
    
    filename = str(uuid.uuid4())
    
    fin_Model.save(filename)

    return {
        'loss':loss,'acc':-acc,
        'status':STATUS_OK,
        'filename':filename
    }
```

```python
trialfin=Trials()

fmin(
    objective,
    space=seq_space,
    algo=tpe.suggest,
    max_evals=10, # 10번의 파라미터 튜닝
    trials=trialfin)
```
